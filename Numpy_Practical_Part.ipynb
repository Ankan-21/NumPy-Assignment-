1. Create a 3x3 NumPy array with random integers between 1 and 100. Then, interchange its rows and columns.


Ans- We can create a 3x3 NumPy array with random integers between 1 and 100, and then interchange its rows and columns (transposing the array). Here's how we can do it:


import numpy as np  

# Create a 3x3 array with random integers between 1 and 100  
array = np.random.randint(1, 101, size=(3, 3))  

# Print the original array  
print("Original Array:")  
print(array)  

# Interchange rows and columns (transpose the array)  
transposed_array = array.T  

# Print the transposed array  
print("\nTransposed Array:")  
print(transposed_array)  

This code uses numpy to generate the random integers and then transposes the array. The .T property of a NumPy array is used to get the transpose of the array, effectively interchanging its rows and columns.

3.Create a 4x4 NumPy array with random float values. Add a border of zeros around it, resulting in a 6x6 array.


Ans- To create a 4x4 NumPy array with random float values and then add a border of zeros around it, we can use NumPy's np.pad function. Here's how we  can do it:


import numpy as np  

# Create a 4x4 array with random float values  
array_4x4 = np.random.rand(4, 4)  

# Print the original 4x4 array  
print("Original 4x4 Array:")  
print(array_4x4)  

# Add a border of zeros around the array  
array_with_border = np.pad(array_4x4, pad_width=1, mode='constant', constant_values=0)  

# Print the new 6x6 array with a border of zeros  
print("\n6x6 Array with Border of Zeros:")  
print(array_with_border)  

Explanation:

np.random.rand(4, 4) generates a 4x4 array with random float values between 0 and 1.
np.pad is used to add a border of zeros around the array. The pad_width=1 argument specifies that one layer of zeros should be added around the original array. mode='constant' with constant_values=0 sets the value of the padding to zero.
The resulting 6x6 array has the original values in the center and a border of zeros surrounding them.

9.Generate a NumPy array of 100 random integers between 0 and 1000. Find and display all prime numbers in
this array.



Ans- Here's how we can generate a NumPy array of 100 random integers between 0 and 1000 and then find and display all prime numbers in that array.


import numpy as np  

# Generate an array of 100 random integers between 0 and 1000  
random_integers = np.random.randint(0, 1001, size=100)  

# Function to check if a number is prime  
def is_prime(num):  
     if num < 2:  
        return False  
     for i in range(2, int(num ** 0.5) + 1):  
        if num % i == 0:  
            return False  
     return True  

# Find all prime numbers in the array  
prime_numbers = [num for num in random_integers if is_prime(num)]  

# Display the results  
print("Random Integers Array:")  
print(random_integers)  

print("\nPrime Numbers:")  
print(prime_numbers)  

Explanation of the Code:

Generate Random Integers: The line np.random.randint(0, 1001, size=100) creates an array of 100 random integers in the range from 0 to 1000.

Prime Checking Function: The is_prime function checks if a number is prime:
It returns False for any number less than 2 (not prime).
For numbers 2 or greater, it checks divisibility from 2 to the square root of that number. If a factor is found, it returns False.
If no factors are found, it returns True (the number is prime).

List Comprehension: It uses a list comprehension to create a list of prime numbers by checking each number in the random integers array with the is_prime function.

Display Output: Finally, it prints the random integers and the list of prime numbers.

Output:
When we run the code, we will see output similar to this (actual values will vary due to randomness):

Random Integers Array:  
[543, 290, 29, 117, 962, 3, 442, ...]  # This will show 100 random integers  

Prime Numbers:  
[3, 29, 131, 383, 547, ...]  # This will show all prime numbers found in the array  

The Random Integers Array contains 100 random integers, and the Prime Numbers section lists all the prime integers found in that array.


10.Create a NumPy array representing daily temperatures for a month. Calculate and display the weekly
averages.



Ans- To create a NumPy array representing daily temperatures for a month and calculate the weekly averages, we can follow this example:


import numpy as np  

# Create a NumPy array representing daily temperatures for a month (30 days)  
# Let's generate random temperatures between -10 and 35 degrees Celsius  
daily_temperatures = np.random.uniform(low=-10, high=35, size=30)  

# Calculate the weekly averages  
# Reshape the array into a shape of (4, 7) to represent 4 weeks  
# Each row will represent a week, and each column will represent a day of the week  
weekly_temperatures = daily_temperatures.reshape(4, 7)  

# Calculate the average for each week  
weekly_averages = np.mean(weekly_temperatures, axis=1)  

# Display the daily temperatures and the weekly averages  
print("Daily Temperatures for the Month:")  
print(daily_temperatures)  

print("\nWeekly Averages:")  
print(weekly_averages)  

Explanation:

Generating Daily Temperatures: The np.random.uniform(low=-10, high=35, size=30) function creates an array of 30 random temperatures between -10 and 35 degrees Celsius.

Reshaping the Array: The reshape(4, 7) method reshapes the array into a 4x7 matrix, where each row represents a week.

Calculating Weekly Averages: The np.mean(..., axis=1) function calculates the mean for each row (week), giving the average temperature for that week.

Displaying Results: The code prints both the daily temperature array and the calculated weekly averages.

Output:
When we run the code, it will display the daily temperatures for the month and the corresponding weekly averages. Since the temperatures are generated randomly, the output will vary each time we run the code. Here's an example of what the output might look like:
